```python
x = np.arange(0, 4*np.pi, .1)   # start,stop,step
y = np.sin(x)

x_tic = list(set([round(val + .5) if val + .5 < 1 else int(val) for val in x]))
x_tix = [val - int(len(x_tic)/2) for val in x_tic]
# y_tic = list(set([round(val + .5) if val + .5 < 1 else int(val) for val in x]))
fig, ax = plt.subplots()
ax.plot(x, y, 'red')
plt.xticks(x_tic)
ax.set_xticklabels(x_tix)



# to set title
plt.title("Oscyloscope run")



plt.grid(True, color = "grey", linewidth =.7, linestyle = "--")


```
